## 3.6 Tabla Resumen de Variables

Tabla consolidada con informaciÃ³n de cada variable:

```python
# Crear tabla resumen
resumen_variables = []

for col in df.columns:
    tipo = 'NumÃ©rica' if col in vars_numericas else 'CategÃ³rica'
    n_unicos = df[col].nunique()
    n_faltantes = df[col].isna().sum()
    pct_faltantes = (n_faltantes / len(df) * 100)
    
    resumen_variables.append({
        'Variable': col,
        'Tipo': tipo,
        'Valores Ãºnicos': n_unicos,
        'Faltantes': n_faltantes,
        '% Faltantes': round(pct_faltantes, 2)
    })

df_resumen = pd.DataFrame(resumen_variables)
print("TABLA RESUMEN DE VARIABLES")
print("â•" * 75)
print(df_resumen.to_string(index=False))
```
